```python
tips=data
tips
```

```python
tips.head()
```

```python
# CATPLOT [Figure Level Function]
#Stripplot
#Swarmplot
sns.catplot(x='day',y='tip',kind='strip',data=tips)
```

```python
sns.catplot(x='day', y='tip', kind='strip', data=tips, hue='day', palette='bright',jitter=.2) # Jitter gives noises to the data.
#In this code, we added the hue='day' parameter to the sns.catplot function, which assigns different colors to each
#unique value in the 'day' column. The palette='bright' argument specifies the color palette to be used for mapping the
#days to colors. The 'bright' palette is a qualitative color palette that provides distinct and easily distinguishable
#colors for categorical data.
```

```python
sns.catplot(x='day', y='tip', kind='swarm',hue='day', data=tips, palette='bright')
```

```python
sns.catplot(x='day', y='tip', kind='swarm',hue='sex', data=tips, palette='bright')
```

```python
# We can do the same things using directly swarm plot
sns.swarmplot(x='day',y='tip',data=tips,hue='day', palette='bright')
```

```python
# Boxplots
# used to summarize the distribution of a continuous variable across different categories or groups. It provides a concise
#and effective way to visualize the median, quartiles, and potential outliers of the data.
sns.boxplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips,hue='sex') # Here,we added one more category i.e sex.
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',hue='sex',data=tips)
```

```python
# Heatmaps
# A heat map is a graphical representation of data where the individual values contained in a matrix are represented as colors.
flights=pd.read_csv('Downloads\\flight.csv')
flights
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5)      # cbar removes the scale.linewidth gives space b/w cells.
```

```python
sns.heatmap(x,cbar=True,linewidth=0.1,annot=True,fmt='d')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='winter')
```

```python
# Jointplot
# joinplot creates a multi_panel figure that shows both the bivariate(or join) relationship between two variable along
# with the univariable(or marginal) distribution of each on separate axes.
# We can use jointplots in the following 4 manners: scatterplots,Hexabinplots,kdeplots,Regplot
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# Distplots
# Distplots are used to do univatiate analysis on a continous/numerical data.
# Distplot combines 3 things in one graph: 1.Histogram, 2.kernel Density Estimation(pdf) 3. Ruggplot

titanic=pd.read_csv('Downloads\\titanic.csv')
titanic.head()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(),inplace=True)
```

```python
sns.distplot(titanic['Age'])
```

```python
sns.distplot(titanic['Age'], kde=False)
```

```python
sns.distplot(titanic['Age'],kde=False,bins=10)
```

```python
#plotting only the KDE
sns.distplot(titanic['Age'],hist=False)  # Kde gives the probability.
```

```python
# Plotting only the Rugplot
sns.distplot(titanic['Age'],hist=False,kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'])
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```
