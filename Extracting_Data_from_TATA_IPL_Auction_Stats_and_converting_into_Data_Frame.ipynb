```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
url="https://ticker.finology.in/"
r=requests.get(url)
```

```python
soup=BeautifulSoup(r.text,"lxml")
soup
```

```python
table=soup.find("table",class_="table table-sm table-hover screenertable")
table
```

```python
headers=table.find_all("th")  # th (Table Header tag), tr (Table Rows tag), th is inside tr.
headers
```

```python
title=[]
```

```python
for i in headers:
    ti=i.text
    print(ti)
```

```python
for i in headers:
    ti=i.text
    title.append(ti)
print(title)
```

```python
df=pd.DataFrame(columns=title)
df
```

```python
rows=table.find_all("tr")
rows
```

```python
for i in rows[1:]:
    data=i.find_all("td")
    row=[tr.text for tr in data]
#     print(row)
    l=len(df)
    df.loc[l]=row
print(df)
```

```python
for i in rows[1:]:
    data = i.find_all("td")
    # Clean the text using .strip() to remove extra whitespace or newline characters
    row = [tr.text.strip() for tr in data]
    l = len(df)
    df.loc[l] = row

print(df)
```

```python
import pandas as pd

# Assuming 'rows' is the data you are scraping and 'df' is the DataFrame
df = pd.DataFrame(columns=["Company", "PriceRs.", "Day HighRs."])  # Creating an empty DataFrame with columns

for i in rows[1:]:
    data = i.find_all("td")
    # Clean the text using .strip() to remove extra whitespace or newline characters
    row = [tr.text.strip() for tr in data]
    # Append the cleaned row to the DataFrame
    df.loc[len(df)] = row

print(df)
```

```python
df.to_csv("Stock_Market_Data.csv")
```
