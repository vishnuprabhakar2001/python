```python
# Calculate the percentage of people with diabetes for each age group
df_diabetes_percentage = df.groupby('Age', as_index=False).apply(
    lambda x: pd.Series({
        'Diabetes_percentage': (x['Diagnosis'].mean()) * 100
    })
)

# Sort the result by 'Age' in descending order
df_diabetes_percentage = df_diabetes_percentage.sort_values(by='Age', ascending=False)

# Display the entire result
d_p=df_diabetes_percentage
d_p
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.lineplot(x='Age',y='Diabetes_percentage',data=d_p,color='r')
# plt.grid()
```

This shows the increase in the percentage of Diabetic people with increase in the Age.
