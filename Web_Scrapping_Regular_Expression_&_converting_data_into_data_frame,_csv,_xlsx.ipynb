```python
url="https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
```

```python
r=requests.get(url)
```

```python
soup=BeautifulSoup(r.text, "lxml")
soup
```

```python
names=soup.find_all("a",class_="title")
names
```

```python
for i in names:
    print(i.text)
```

```python
#using list []
pduct_list=[]
```

```python
for i in names:
    n=i.text
    pduct_list.append(n)
pduct_list
```

```python
prices=soup.find_all("h4",class_="price float-end card-title pull-right")
prices
```

```python
price_list=[]
```

```python
for i in prices:
    pr=i.text
    price_list.append(pr)
price_list
```

```python
desc=soup.find_all("p",class_="description card-text")
desc
```

```python
desc_list=[]
```

```python
for i in desc:
    d=i.text
    desc_list.append(d)
desc_list
```

```python
reviews=soup.find_all("p",class_="review-count float-end")
reviews
```

```python
review_list=[]
```

```python
for i in reviews:
    n=i.text
    review_list.append(n)
review_list
```

```python
#*** Converting the given list into DataFrame from unstructured to structured.
df=pd.DataFrame({"Product_name":pduct_list,"Prices":price_list,"Description":desc_list,"Reviews":review_list})
df
```

```python
# Converting into csv
df.to_csv("Product_details.csv")
#If you also want to exclude the index column from being saved in the CSV file, you can add the index=False argument:

# df.to_csv("Product_details.csv", index=False)

#This will save only the DataFrame's data without the row numbers (index).
```

```python
# Converting into Excel
df.to_excel("Product_details.xlsx")
```
